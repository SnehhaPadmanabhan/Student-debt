```
import pandas as pd
data_xls = pd.read_excel('data.xls', 'Digest 2017 Table 105.30', index_col=None)
data_xls.to_csv('data.csv', encoding='utf-8')
import pandas as pd
dataframe = pd.read_csv("data.csv").iloc[:, 1:]
print(dataframe.head(10))
```

```
print(dataframe.columns)
```

```
import matplotlib.pyplot as plt
plt.plot(dataframe.Years, dataframe.Total, label = 'Total')
plt.plot(dataframe.Years, dataframe.Public, label = 'Public institutions', color  = 'Red')
plt.plot(dataframe.Years, dataframe.Private, label= 'Private institutions', color= 'Green')
plt.xlabel('Years from 1940-2019')
plt.ylabel('Number of enrollments')
plt.title('Number of enrollments')
plt.legend()
plt.xticks(dataframe.Years,  ' ')
plt.savefig('enrollments.png')
plt.show()
```
