```
import pandas as pd
salary_xls = pd.read_excel('faculty_salary.xlsx', 'Sheet1', index_col = None)
salary_xls.to_csv('salary.csv', encoding='utf-8')
dataframe = pd.read_csv("salary.csv").iloc[:, 1:]
print(dataframe.head(10))
```

```
dataframe.columns
```

```
dataframe.rename(columns = {'Associate professors':'Associate_professors', 'Assistant professors':'Assitant professors'}, inplace = True)
```

```
dataframe.rename(columns = {'Assitant professors':'Assistant_professors'}, inplace = True)
```

```
import matplotlib.pyplot as plt
plt.plot(dataframe.Year, dataframe.Professors, label = 'Professors', color = 'Black')
plt.plot(dataframe.Year, dataframe.Associate_professors, label = 'Associate professors', color  = 'Red')
plt.plot(dataframe.Year, dataframe.Assistant_professors, label= 'Assistant professors', color= 'Green')
plt.plot(dataframe.Year, dataframe.Instructors, label= 'Instructors', color= 'Blue')
plt.plot(dataframe.Year, dataframe.Lecturers, label= 'Lecturers', color= 'Orange')
plt.plot(dataframe.Year, dataframe.Unranked, label= 'Unranked', color= 'Yellow')
plt.xlabel('Years from 2003-2019')
plt.ylabel('Salary in dollars')
plt.title('Average salary of faculty in post secondary institutions')
plt.legend()
plt.xticks(dataframe.Year,  ' ')
plt.savefig('salary.png')
plt.show()
```
