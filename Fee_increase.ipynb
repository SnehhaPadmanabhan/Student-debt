```
import pandas as pd
data_xls = pd.read_excel('all_fee.xls', 'all_fee', index_col=None)
data_xls.to_csv('total_fee.csv', encoding='utf-8')
```

```
import pandas as pd
dataframe = pd.read_csv("total_fee.csv").iloc[:, 1:]
print(dataframe.head(10))
```

```
dataframe.rename(columns = {'Unnamed: 0.1':'Years'}, inplace = True)
```

```
print(dataframe.columns)
```

```
import matplotlib.pyplot as plt
dataframe.rename(columns = {'Total ':'Total'}, inplace = True)
plt.plot(dataframe.Years, dataframe.Total)
plt.show()
```

```
plt.plot(dataframe.Years, dataframe.Total, label = 'Total', color = 'Black')
plt.plot(dataframe.Years, dataframe.Tution, label = 'Tution fee')
plt.plot(dataframe.Years, dataframe.Room, label = 'Room fee', color  = 'Red')
plt.plot(dataframe.Years, dataframe.Board, label= 'Boarding fee', color= 'Green')
plt.xlabel('Years from 1964-2017')
plt.ylabel('Amount in dollars(current)')
plt.title('All institutions')
plt.legend()
plt.xticks(dataframe.Years,  ' ')
plt.savefig('all_fee.png')
plt.show()
```

```
data_xls = pd.read_excel('private_fee.xls', 'Digest 2017 Table 330.10', index_col=None)
data_xls.to_csv('private.csv', encoding='utf-8')
dataframe_private = pd.read_csv("private.csv").iloc[:, 1:]
# print(dataframe_private.head(10))
print(dataframe_private.columns)
plt.plot(dataframe_private.Years, dataframe_private.Total, label = 'Total', color = 'Black')
plt.plot(dataframe_private.Years, dataframe_private.Tution, label = 'Tution fee')
plt.plot(dataframe_private.Years, dataframe_private.Room, label = 'Room fee', color  = 'Red')
plt.plot(dataframe_private.Years, dataframe_private.Board, label= 'Boarding fee', color= 'Green')
plt.xlabel('Years from 1964-2017')
plt.ylabel('Amount in dollars(current)')
plt.title('Private institutions')
plt.legend()
plt.xticks(dataframe_private.Years,  ' ')
plt.savefig('private_fee.png')
plt.show()
```

```
data_xls = pd.read_excel('public_fee.xls', 'Digest 2017 Table 330.10', index_col=None)
data_xls.to_csv('public.csv', encoding='utf-8')
dataframe_public = pd.read_csv("public.csv").iloc[:, 1:]
# print(dataframe_private.head(10))
print(dataframe_public.columns)
plt.plot(dataframe_public.Years, dataframe_public.Total, label = 'Total', color = 'Black')
plt.plot(dataframe_public.Years, dataframe_public.Tution, label = 'Tution fee')
plt.plot(dataframe_public.Years, dataframe_public.Room, label = 'Room fee', color  = 'Red')
plt.plot(dataframe_public.Years, dataframe_public.Board, label= 'Boarding fee', color= 'Green')
plt.xlabel('Years from 1964-2017')
plt.ylabel('Amount in dollars(current)')
plt.title('Public institutions')
plt.legend()
plt.xticks(dataframe_public.Years,  ' ')
plt.savefig('public_fee.png')
plt.show()
```

```
plt.plot(dataframe.Years, dataframe.Total, label = 'All institutions', color = 'Green')
plt.plot(dataframe_public.Years, dataframe_public.Total, label = 'Public', color = 'Blue')
plt.plot(dataframe_private.Years, dataframe_private.Total, label = 'Private', color = 'Red')
plt.xlabel('Years from 1964-2017')
plt.ylabel('Amount in dollars(current)')
plt.title('Comparison graph')
plt.legend()
plt.xticks(dataframe_public.Years,  ' ')
plt.savefig('comparison.png')
plt.show()
```
