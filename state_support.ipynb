```
import pandas as pd
support_xls = pd.read_excel('state_support.xlsx', 'Sheet1', index_col=None)
support_xls.to_csv('support.csv', encoding='utf-8')
dataframe = pd.read_csv("support.csv").iloc[:, 1:]
print(dataframe.head(10))
```

```
import matplotlib.pyplot as plt
plt.plot(dataframe.Year, dataframe.Amount, label = 'Amount')
plt.xlabel('Years from 2000-2018')
plt.ylabel('Amount in constant dollars')
plt.title('State Support for Higher Education per Full-Time Equivalent Student (Dollars)')
plt.legend()
plt.xticks(dataframe.Year,  ' ')
plt.savefig('support.png')
plt.show()
```
